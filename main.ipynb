# The code below prints the models output

```python

print("Generating new verations of the Original Monet painting using Random noise as an input to the first scale of the trained model\n\n")
#Input Painting folder
input_path = "./Input/Images/"

# Output folder path
plot_path = "./Output/IU/RandomSamples/monet96/gen_start_scale=0,alpha=1.000000/"
# reading original panting
Original = mpimg.imread(input_path + "monet96.jpg")
# showing image
plt.imshow(Original)
plt.axis('off')
plt.title("Original")

# create figure
fig = plt.figure(figsize=(17, 17))
#fig = plt.figure()
# setting values to rows and column variables
rows = 1
columns = 4

# reading images
Image1 = mpimg.imread(plot_path + "0.png")
Image2 = mpimg.imread(plot_path + "1.png")
Image3 = mpimg.imread(plot_path + "2.png")
Image4 = mpimg.imread(plot_path + "3.png")

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)

# showing image
plt.imshow(Image1)
plt.axis('off')
plt.title("01")

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)

# showing image
plt.imshow(Image2)
plt.axis('off')
plt.title("02")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)

# showing image
plt.imshow(Image3)
plt.axis('off')
plt.title("03")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)

# showing image
plt.imshow(Image4)
plt.axis('off')
plt.title("04")

```

# Inject photo into a SingGAN model

```python
!python paint2image.py --input_name monet134.jpg --ref_name photo2668.jpg --paint_start_scale 3  --scale_factor 0.3 --alpha 1 --noise_amp 1.0 --input_dir ./Input/Images/ --ref_dir ./Input/photos --out ./Output/Injection/
```

# Print injetion example

```python
print("Example of photo injection to style transfer the Modet painting\n\n")
#Input Painting folder
input_path = "./Input/Images/"
photo_path = "./Input/photos/"
# Output folder path
plot_path = "./Output/Injection/Paint2image/monet134/photo2668_out/start_scale=3.png"

# create figure
fig = plt.figure(figsize=(17, 17))
#fig = plt.figure()
# setting values to rows and column variables
rows = 1
columns = 3

# reading images
monet_original = mpimg.imread(input_path + "monet134.jpg")
photo = mpimg.imread(photo_path + "photo2668.jpg")
photo_injected = mpimg.imread(plot_path)

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)

# showing image
plt.imshow(monet_original)
plt.axis('off')
plt.title("Monet Paintings")

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)

# showing image
plt.imshow(photo)
plt.axis('off')
plt.title("Photo")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)

# showing image
plt.imshow(photo_injected)
plt.axis('off')
plt.title("photo to Monet painting")
```
